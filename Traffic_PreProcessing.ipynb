### Exploratory Data Analysis Summary
Our data looks at eight different variables to try and predict the traffic flow. Traffic flow is a categorical variable with classes normal, heavy, high, and low. The data looks at the number of different types of vehicles on a given road at 15 minute intervals through the day, spanning one month, and then has an assigned traffic situation to each of those 15 minute periods. The predictors in the dataset are both categorical and quantitative. Time and Weekday are the two categorical predictors. Since we are creating a boosting model, we will not need to dummify these variables. However, if we were to compare the accuracy of our boosting model to a linear regression model, we would need to transform the variables. 

It makes sense to do variable selection for this model since many of the quantitative predictors are highly correlated so we want to avoid multicollinearity in our model. This result makes sense since 1) the Total predictor is the sume of the CarCount, BikeCount, BusCount, and TruckCount predictors and 2) the response variable doesn't differentiate betwen the type of vehicle on the road to determine what class is assigned to the 15 minute window. So, we will take out the various individualized counts and keep in the total vehicle count. 

We note that for the response variable, 56.1% of the data belongs to the "normal" class, which means we have an imbalanced classification problem. We know that an issue with a model trying to classify traffic flow will be that it will have a hard time predicting the high and low classes, which have a prior pribability of 10.8% and 10.2%, respectively.
